## 3. Consultas básicas con JOIN

```python
#Import de la libreria de Sqlite.

import sqlite3

#Consulta para juntar las dos bases de datos por el punto en comun.

id = 1
with sqlite3.connect("empresa.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT Empleados.nombre, Empleados.puesto, Departamentos.nombre FROM Empleados JOIN Departamentos ON Empleados.departamento_id = Departamentos.id")
    resultado = cur.fetchall()
    for i in resultado:
        print(f"Empleado {id}: {i[0]} | Puesto: {i[1]} | Departamento: {i[2]}")
        id += 1

#Consulta para sacar los salarios que sea superior a la media de los salarios.

with sqlite3.connect("empresa.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT nombre, salario FROM Empleados WHERE salario > (SELECT AVG(salario) FROM Empleados)")
    filtro = cur.fetchall()
    for i in filtro:
        print(f"Nombre: {i[0]} | Salario: {i[1]}")
```
