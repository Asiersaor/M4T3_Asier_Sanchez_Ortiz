## 3. Consultas básicas con JOIN

```python
#Import de la libreria de Sqlite.

import sqlite3

#Consulta para juntar las dos bases de datos por el punto en comun.

id = 1
with sqlite3.connect("empresa.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT Empleados.nombre, Empleados.puesto, Departamentos.nombre FROM Empleados JOIN Departamentos ON Empleados.departamento_id = Departamentos.id")
    resultado = cur.fetchall()
    for i in resultado:
        print(f"Empleado {id}: {i[0]} | Puesto: {i[1]} | Departamento: {i[2]}")
        id += 1

#Consulta para sacar los salarios que sea superior a la media de los salarios.

with sqlite3.connect("empresa.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT nombre, salario FROM Empleados WHERE salario > (SELECT AVG(salario) FROM Empleados)")
    filtro = cur.fetchall()
    for i in filtro:
        print(f"Nombre: {i[0]} | Salario: {i[1]}")
```

## 4. Actualización de datos

```python
#Import de la libreria.

import sqlite3

#Incremento del 15% para los empleados que tengan una fecha de contratacion menor de 2020.

with sqlite3.connect("empresa.db") as conn:
    cur = conn.cursor()
    cur.execute("UPDATE Empleados SET salario = salario * 1.15 WHERE fecha_contratacion < '2020-01-01'")

#Bajada del 10% para los empleados cuyo departamento tenga perdidas.

with sqlite3.connect("empresa.db") as conn:
    cur = conn.cursor()
    cur.execute("SELECT SUM(Empleados.salario), Departamentos.nombre FROM Empleados JOIN Departamentos ON Empleados.departamento_id = Departamentos.id GROUP BY Departamentos.id")
    cur.execute("UPDATE Departamentos SET presupuesto = presupuesto * 0.90 WHERE presupuesto < (SELECT SUM(Empleados.salario) FROM Empleados WHERE Empleados.departamento_id = Departamentos.id)")
```
